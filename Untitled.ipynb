```python
import zipfile

with zipfile.ZipFile("Sample - Superstore.csv.zip", 'r') as zip_ref:
    zip_ref.extractall("data")
```

```python

 import pandas as pd

df = pd.read_csv("data/Sample - Superstore.csv", encoding='latin1')
df.head()
```

```python
df.info()
df.columns
```

```python
# Convert date
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Remove duplicates
df = df.drop_duplicates()

# Check nulls
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

# Extract month-year
df['Month-Year'] = df['Order Date'].dt.to_period('M')

monthly_sales = df.groupby('Month-Year')['Sales'].sum()

plt.figure()
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Time")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
#Sales peak during year-end (Nov–Dec)
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

plt.figure()
category_sales.plot(kind='bar')
plt.title("Sales by Category")
plt.ylabel("Sales")
plt.show()
#Technology category generates highest revenue
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

print(top_products)
#Few products contribute majority of revenue
```

```python
low_products = df.groupby('Product Name')['Sales'].sum().sort_values().head(10)

print(low_products)
```

```python
plt.figure()
plt.scatter(df['Sales'], df['Profit'])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
#Some products have high sales but low/negative profit
```

```python


region_sales = df.groupby('Region')['Sales'].sum()

plt.figure()
region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.show()
```

```python
#Focus marketing on top-performing category
#Increase inventory before peak season
#Discontinue or improve low-performing products
#Optimize pricing for high-sales but low-profit items
```
